```python
def plot_box_on_map(plot_ax, lonW, lonE, latN, latS,colorline,linewidth,exact,num_pts=20, delta=0.5):
        
    # Check if exact is equal to 'yes'
    if exact == 'yes':
        lonW = lonW
        lonE = lonE
        latN = latN
        latS = latS
    else:
        lonW = lonW-delta
        lonE = lonE+delta
        latN = latN+delta
        latS = latS-delta
    
    line_lons = []
    line_lats = []
    
    # west edge
    line_lons.append(np.ones(num_pts)*lonW)
    line_lats.append(np.linspace(latS,latN,num_pts))
    
    # north edge
    line_lons.append(np.linspace(lonW,lonE,num_pts))
    line_lats.append(np.ones(num_pts)*latN)
    
    # east edge
    line_lons.append(np.ones(num_pts)*lonE)
    
    line_lats.append(np.linspace(latS,latN,num_pts))
    # south edge

    line_lons.append(np.linspace(lonW,lonE,num_pts))
    line_lats.append(np.ones(num_pts)*latS)
    # plot 4 lines on map
    color=colorline
    for i in range(4):
        plot_ax.plot(line_lons[i], line_lats[i],
             color=colorline, linewidth=linewidth, marker='.', markersize=1,
             transform=crs.PlateCarree(), zorder=100)
```

# Mask for 2D surface area in a box

```python
# load the ecco grid into memory
grid.load();
```

```python
# select the surface index
depthINDEX=0 #  is surface

# pull the land mask
land_mask=grid.maskC.isel(k=depthINDEX).copy(deep=True)
land_mask_nan = land_mask.copy(deep=True)
land_mask_nan = land_mask_nan.where(land_mask_nan == True)

# define box boundaries
lonW = -130; lonE = 0; latN = 45; latS = -20

lat_mask = np.where(np.logical_and(grid.YC > latS, grid.YC < latN), 1, 0)
lon_mask = np.where(np.logical_and(grid.XC > lonW, grid.XC < lonE), 1, 0)
latlon_mask = lat_mask * lon_mask * land_mask

latlon_mask_da = grid.XC.copy(deep=True)*0
latlon_mask_da.values[:] = latlon_mask

# multiple the grid cell area with the land mask
area_mask=grid.rA*latlon_mask_da

# calculate the total area of wet cells in tile 11
total_area_mask=area_mask.sum(dim=['i','j','tile']).compute()

# the mask is defined by the a(i,j)/total_area_mask
# each grid cell is weighted by its fraction of the total area
area_mask_coef=area_mask/total_area_mask

# make a new DataArray object, use grid.XC as the template for coordiantes, etc.
global_2D_mask = grid.XC.copy(deep=True)*0
# populate the values with the area-weighted mask values calculated above
global_2D_mask.values[:] = area_mask_coef[:]
# give some helpful name
global_2D_mask.name = 'custom_2D_mask'
global_2D_mask.attrs['long_name'] = 'custom area weighted 2D mask of wet cells in tile 11'

# verify the global sum of the mask is 1
print(f'global sum of area-weighted mask values is {global_2D_mask.sum().values}')
ecco.plot_tiles(global_2D_mask, layout='latlon', rotate_to_latlon=True, fig_size=5, show_colorbar=True);


plt.figure(figsize=[10,5])
# verify the global sum of the mask is 1
X = ecco.plot_proj_to_latlon_grid(grid.XC, grid.YC, 
                                  global_2D_mask*land_mask_nan,dx=1,dy=1, 
                                  show_colorbar=True,show_grid_labels=True);
plot_box_on_map(X[1], lonW, lonE, latN, latS,
                colorline='red',linewidth=2.5,exact='yes')
```

# Save mask in binary (Big Endian Byte Order) and netcdf

```python
global_2D_mask_compact = ecco.llc_tiles_to_compact(global_2D_mask)
```

```python
name_mask='global_2D_mask_box'
# for binary 
from pathlib import Path
file_path_compact = Path(name_mask+'.bin')
with open(file_path_compact, "wb") as file:
    global_2D_mask_compact.astype('>f4').tofile(file)
# for netcdf
file_path_netcdf = Path(name_mask+'.nc')
with open(file_path_netcdf, "wb") as file:
    global_2D_mask.to_netcdf(file)
```

```python
# read the mask in netcdf format
test_nc = xr.open_dataset(file_path_netcdf)
print(test_nc.data_vars)
ecco.plot_tiles(test_nc.custom_2D_mask,layout='latlon', 
                rotate_to_latlon=True, fig_size=5, 
                show_colorbar=True);

plt.figure(figsize=[10,5])
X=ecco.plot_proj_to_latlon_grid(test_nc.XC, test_nc.YC,
                                  test_nc.custom_2D_mask,dx=1,dy=1, 
                                  show_colorbar=True,show_grid_labels=True);
plot_box_on_map(X[1], lonW, lonE, latN, latS,
                colorline='red',linewidth=2.5,exact='yes')
```
